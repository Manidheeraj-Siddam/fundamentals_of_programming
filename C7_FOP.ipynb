```python
CRLF = '\n'

def make_pyramid(n: int) -> str:
    pyramid = []
    i = n
    j = 0
    while i >= 1:
        pyramid.append(" " * j + "* " * i)
        i -=1
        j +=1
    return CRLF.join(pyramid)

print(make_pyramid(5))
```

```python
B = '*'
W = ' '
def pad_with_blacks(grid: list[str]) -> list[str]:
    padded_grid = []
    for row in grid:
        padded_grid += [B + row + B]
    padded_grid = [B * len(padded_grid[0])] +\
                   padded_grid +\
                   [B * len(padded_grid[0])]
    return padded_grid

def is_start_clue_down(grid: list[str], row: int, col: int) -> bool:
    return (grid[row - 1][col], grid[row][col], grid[row + 1][col]) ==(B, W, W)

def is_start_clue_across(grid: list[str], row: int, col: int) -> bool:
    return (grid[row][col - 1], grid[row][col], grid[row][col + 1]) == (B, W, W)

def number(grid: list[str]) -> list[tuple[int, int, int]]:
    grid = pad_with_blacks(grid)
    numberings = []
    current_number = 1
    for row in range(1, len(grid) - 1):
        for col in range(1, len(grid[0]) - 1):
            if is_start_clue_across(grid, row, col) or is_start_clue_down(grid, row, col):
                numberings += [(row-1, col-1, current_number)]
                current_number += 1

    return numberings
```
