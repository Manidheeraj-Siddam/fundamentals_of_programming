```python
import math

class KNearestNeighbors:
    def __init__(self):
        pass

    def euclidean_distance(self, point1: float, point2: float) -> float:
        return math.sqrt(sum((a - b) ** 2 for a, b in zip(point1, point2)))

    def k_nearest_neighbors(self, points, k):
        reference_point = (0, 0)

        sorted_points = sorted(points, key=lambda p: self.euclidean_distance(reference_point, p))

        k_nearest = sorted_points[:k]

        return k_nearest
```

```python
from __future__ import annotations

class Point:
    def __init__(self, x: int, y: int) -> None:
        self.x = x
        self.y = y

class Circle:
    PI = 3.14

    def __init__(self, radius: int, center: Point) -> None:
        self.radius = radius
        self.center = center

    def area(self) -> float:
        return Circle.PI * (self.radius ** 2)

    def contains_point(self, point: Point) -> bool:
        distance_squared = (point.x - self.center.x)**2 + (point.y - self.center.y)**2
        return distance_squared <= self.radius**2

    def __eq__(self, other: Circle) -> bool:
      return self.radius == other.radius and self.center == other.center

    def __repr__(self) -> str:
        h, k = self.center.x, self.center.y
        return f"(x - {h})^2 + (y - {k})^2 = {self.radius**2}"


center_point = Point(2, 3)
my_circle = Circle(5, center_point)
my_circle
```

```python
class Circle:

    INSIDE = "inside"
    ON = "on"
    OUTSIDE = "outside"

    def __init__(self, centre: Point, radius: float) -> None:
        self.centre = centre
        self.radius = radius

    def classify (self, point: Point2D) -> str:
        if point.distance(self.centre) > self.radius:
            return Circle.OUTSIDE
        elif point.distance(self.centre) == self.radius:
            return Circle.ON
        elif point.distance(self.centre) < self.radius:
            return Circle.INSIDE

    def scale(self, factor: float) -> None:
        self.radius *= factor

    def scaled(self, factor: float) -> Circle:
        return Circle(self.centre, self.radius * factor)

    def area(self) -> float:
        return math.pi * self.radius ** 2

    def perimeter(self) -> float:
        return 2 * math.pi * self.radius

    def num_intersections(self, line: Line2D) -> int:
        distance = line.distance(self.centre)
        if distance > self.radius:
            return 0
        elif distance == self.radius:
            return 1
        elif distance < self.radius:
            return 2

    def __eq__(self, other) -> bool:
        return self.centre == other.centre and self.radius == other.radius

    def __repr__(self) -> str:
        return f'Centre: {self.centre}, Radius: {self.radius}'

    def __str__(self) -> str:
        return f'''Circle Equation: (y - {self.centre.y})^2 + (x - {self.centre.x})^2 = {self.radius}^2
        {self.__repr__()}'''
```

```python
class Square:

    def __init__(self, value: int) -> None:
        self.value = value
        self.marked = False

    def __repr__(self) -> str:
        if self.marked:
            return f'[{self.value}]'
        else:
            return f'{self.value}'
```

```python
import random

class Grid:
    def __init__(self, size: int) -> int:
        self.size = size
        nums = list(range(1, size * size + 1))
        random.shuffle(nums)
        square_sequence = [Square(i) for i in nums]
        self.squares = [square_sequence[a-size : a]
                        for a in range(size, size * size + 1, size)]

    def mark(self, row, col) -> None:
        self.squares[row][col].marked = True

    def bingo(self) -> bool:
        for row in self.squares:
            if all(square.marked for square in row):
                return True
        for col in range(self.size):
            if all(self.squares[row][col].marked for row in range(self.size)):
                return True
        if all(self.squares[i][i].marked for i in range(self.size)) or \
           all(self.squares[i][self.size - 1 - i].marked for i in range(self.size)):
            return True

        return False

    def play(self):
        num = list(range(1, self.size*self.size + 1))
        random.shuffle(num)
        for ch in num:
           i,j = self.find_indices(ch)
            self.mark(i, j)
            self.__str__()



    def find_indices(self ,ch: int):
            for i, row in enumerate(self.squares):
                for j, element in enumerate(row):
                    if element == ch:
                        return i,j


    def __str__(self) -> str:
        return '\n'.join(['|'.join
                 ([str(num) for num in row])
                                for row in self.squares])
```
