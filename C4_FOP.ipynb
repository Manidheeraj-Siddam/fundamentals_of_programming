```python
def is_prime(n: int) -> bool:
    div = 2
    while div <= n ** 0.5:
        if n % div == 0:
            return False
        div += 1
    return True

def num_to_digits(n: int) -> list[int]:
    return [int(k) for k in str(n)]

def digits_to_num(digs: list[int]) -> int:
    return int(''.join([str(k) for k in digs]))

def next_cyclical_permute(num: int) -> int:
    digs = num_to_digits(num)
    digs = [digs[-1]] + digs[:-1]
    return digits_to_num(digs)

def cyclical_permutations(num: int) -> list[int]:
    out = []
    while num not in out:
        out += [num]
        num = next_cyclical_permute(num)
    return out

def is_circular_prime(num: int) -> bool:
    return all([is_prime(k) for k in cyclical_permutations(num)])

def circular_primes(num_digits: int) -> list[int]:
    out = []
    current = 10 ** (num_digits - 1)
    while current < 10 ** num_digits:
        if is_circular_prime(current) and not any([k in out for k in cyclical_permutations(current)]):
            out += [current]
        current += 1
    return out

circular_primes(5)
```

```python
def is_vowel(ch: str) -> bool:
    return ch in ('a', 'e', 'i', 'o', 'u')

def vowels(sentence: str) -> tuple:
    sentence = sentence.lower()
    vowel_list = [ch for ch in sentence if is_vowel(ch)]
    return tuple(vowel_list)

vowels("there are so many")

```

```python
def lower_decoder(ch: str) -> str:
    return chr((ord(ch) - ord('a') - 13) % 26 + ord('a'))

def upper_decoder(ch: str) -> str:
    return chr((ord(ch) - ord('A') - 13) % 26 + ord('A'))

def rot13(input_str: str) -> str:
    out = ""
    for ch in input_str:
        if ch.islower():
            out += lower_decoder(ch)
        elif ch.isupper():
            out += upper_decoder(ch)
        else:
            out += ch
    return out

```

ROMAN TO INTEGER

```python
ROMAN = {"I":1,"V":5,"X":10,"L":50,"C":100,"D":500,"M":1000}

def roman_to_int(s: str) -> int:
  number = 0
  for i in range(len(s)-1):
      if ROMAN[s[i]] < ROMAN[s[(i+1)]]:
          number -= ROMAN[s[i]]
      else:
          number += ROMAN[s[i]]
  return number + ROMAN[s[-1]]

roman_to_int("XIV")
```

INTEGER TO ROMAN

```python
ROMAN_DICT = {
        1000: "M", 900: "CM", 500: "D", 400: "CD",
        100: "C", 90: "XC", 50: "L", 40: "XL",
        10: "X", 9: "IX", 5: "V", 4: "IV",
        1: "I"
    }

def int_to_roman(num):
    roman_numeral = ''
    for value, symbol in ROMAN_DICT.items():
        while num >= value:
            roman_numeral += symbol
            num -= value

    return roman_numeral

int_to_roman(456)
```
