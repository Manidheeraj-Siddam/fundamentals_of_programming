```python
def pattern_pyramid(n: int) -> None:
  for i in range(1,n+1):
    print(f'{i:3d}' * i)

pattern_pyramid(10)
```

```python
size = 7
CRLF = '\n'

def line(n: int) -> str:
    return f'{n:3d}' * n

def make_pyramid(n: int) -> list[str]:
    pyramid = []
    for i in range(1, n + 1):
        pyramid.append(line(i))
    return pyramid

def format_pyramid(pyramid: list[str]) -> str:
    return CRLF.join(pyramid)

print(format_pyramid(make_pyramid(4)))
```

```python
def generate_triangle(rows):
    for i in range(1, rows + 1):
        for j in range(1, i + 1):
            print(i * j, end = " ")
        print()

generate_triangle(6)
```

```python
def print_pattern(n) -> str
    current_number = 1

    for i in range(1, n + 1):
        for j in range(1, i + 1):
            if(current_number > n):
              return
            print(current_number, end=" ")
            current_number += 1

        print()

print_pattern(10)
```

```python
e = 0.00001
FIRST_GUESS = 1.0

def good_enough(guess, n):
  return abs(guess * guess - n) < e

def improve(guess, n):
  return (guess + n / guess) / 2

def sqrt(n):
  current_guess = FIRST_GUESS
  while not good_enough(current_guess, n):
     current_guess = improve(current_guess, n)
  return current_guess

sqrt(2)
```

```python
NUMS_LETTERS = [" ", ".,#*", "abc", "def", "ghi", "jkl", "mno", "pqrs", "tuv", "wxyz"]

def t9_repr(ch: str) -> int:
    for seq in NUMS_LETTERS:
        if ch in seq:
            return NUMS_LETTERS.index(seq)

def to_t9(sentence: str) -> list[int]:
    return [t9_repr(ch) for ch in sentence]

def num_to_digits(num: int) -> list[int]:
    return [int(ch) for ch in str(num)]

def digits_to_num(digits: list[int]) -> int:
    return int(''.join([str(dig) for dig in digits]))
```

```python
def all_decodings(t9_encoding: list[int]) -> str:
    if t9_encoding == []:
        return [""]
    else:
        return [prev + ch for prev in all_decodings(t9_encoding[1:]) for ch in NUMS_LETTERS[t9_encoding[0]]]
```
